```python
#check outlier in price_per_sqft column and remove it
data['price_per_sqft'].describe()
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        gen_df = subdf[(subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m + st))]
        df_output = pd.concat([df_output, gen_df], ignore_index=True)
    return df_output

data = remove_outliers_sqft(data)
data.describe()
```

```python
data.head(3)
```

```python
#now check outlier in bhk and remove it
def bhk_outlier_removal(df):
    exclude_indices = np.array([])
    for location,location_df in df.groupby('location'):
        bhk = {}
        for BHK,BHK_df in location_df.groupby('BHK'):
            bhk[BHK] = {
                'mean' : np.mean(BHK_df.price_per_sqft),
                'std' : np.std(BHK_df.price_per_sqft),
                'count': BHK_df.shape[0]
            }
        for BHK,BHK_df in location_df.groupby('BHK'):
            stat = bhk.get(BHK-1)
            if stat and stat['count']>5:
                exclude_indices = np.append(exclude_indices, BHK_df[BHK_df.price_per_sqft<(stat['mean'])].index.values)
    return df.drop(exclude_indices, axis = 'index')
```

```python
data = bhk_outlier_removal(data)
```

```python
x = data.drop(columns =['price'])
y = data['price']
```

```python
#importing python library sklearn for Ml
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
#20% of data use to testing and remaining  used for traning.
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2 , random_state = 0)
```

```python
#number of data from dataset for testing.
print(x_test.shape)
print(y_test.shape)
```

```python
#number of data for training...
print(x_train.shape)
```

# Now applying Linear Regression

```python
column_trans = make_column_transformer((OneHotEncoder(sparse_output = False),['location']),remainder = 'passthrough',force_int_remainder_cols=False)
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred_lr)
```

# Now apllying Lasso

```python
lasso = Lasso()
pipe = make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lasso = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred_lasso)
```

# Now Applying Ridge

```python
ridge = Ridge()
pipe = make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_ridge = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred_ridge)
```

```python
#compare LR,Lasso,Ridge accuracy
print("No Regularization(LR) accuracy :",r2_score(y_test,y_pred_lr))
print("Lasso Accuracy :",r2_score(y_test,y_pred_lasso))
print("Ridge Accuracy:",r2_score(y_test,y_pred_ridge))
```
